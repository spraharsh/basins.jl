### 1D Potential Landscapes

Plots for 1D potential landscapes

```julia

# minima locations
loc = [0.7, 2.5, 4, 4.2]

"""
Continuous and differentiable function that has a minimum at zero, and defined from (-1,1)
"""
function minima_piece(x)
    if abs(x) < 1
        return -(1 - x^2)^4
    else
        return 0
    end
end





pot(x) = 0.8minima_piece(x-loc[1]) + minima_piece((x-loc[2])/2) + 0.7minima_piece(x-loc[3]) 


range = 0:0.02:4.6
#plot(range, minima_piece.(range), label = "minima piece")

plot(range, pot.(range), label = "potential", xaxis=false, yaxis=false, legend=false, size=(800, 600), axis = ([], false), linewidth=2, color=:black)
```
